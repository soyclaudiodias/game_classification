## üîπ Importa√ß√µes

## üîπ Carregando o Modelo Zero-Shot
Aqui carregamos o modelo **distilbart-mnli**, uma vers√£o otimizada que permite classificar textos nos g√™neros definidos.

```python
candidate_labels = ["Action", "Adventure", "RPG", "Shooter", "Strategy", "Puzzle", "Horror"]

classifier_llm = pipeline(
    "zero-shot-classification",
    model="valhalla/distilbart-mnli-12-1"
)
```

## üîπ Testando o Modelo
